# **Data Analysis Life Cycle**

5 Steps Data Analysis infographic.jpg

1. Business Knowledge
2. Data Preparation
3. Data Cleaning and Preprocessing
4. Exploratory Data Analysis
5. Result and Data Visualization


## 1.Business Knowledge
Understand your Data
Define the questionWhat business problem are you trying to solve? Frame it as a question to help you focus on finding a clear answer.

## 2. Data Preparation
Collect, Merge and Prepare your data. Create a strategy for collecting data. Which data sources are most likely to help you solve your business problem?

## 3. Data Cleaning and Preprocessing
1. Removing major errors, duplicates, and outliers
> all of which are inevitable problems when aggregating data from numerous sources.
2. Removing unwanted data points
> extracting irrelevant observations that have no bearing on your intended analysis.
3. Bringing structure to your data
> general housekeeping, i.e. fixing typos or layout issues, which will help you map and manipulate your data more easily.
4. Filling in missing data
> as youre tidying up, you might notice that important data are missing. Once youve identified gaps, you can go about filling them.

## 4. Analysing Data
Carry out various analyses to obtain insights. Focus on the four types of data analysis: descriptive, diagnostic, predictive, and prescriptive.

## 5. Result and Data Visualization
How best can you share your insights and recommendations? A combination of visualization tools and communication is key.
