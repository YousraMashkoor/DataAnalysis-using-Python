# Task:01

Choose an appropriate graph to display the change in sales of each category throughout the year. Display each sale inside a separate graph; *Hint: Make use of subplots

```python
# Function to add value on top of bars in a bargraph
def autolabel(rects, ax, color='r', rot=0):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        # '%g' % is done to drop trainling zeros from decimal position
        ax.annotate('%g' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color=color, fontsize=13, rotation=rot)
```

```python
fig, ax = plt.subplots(4,3,figsize=(15,15))
plt.tight_layout()
fig.suptitle('Trend of Sales for each Product Category (Furniture, Office Supplies, Technology) from 2014 - 2017', fontsize=18)
fig.subplots_adjust(top=0.95)
#d = {'0':cats_data_2014, 1:cats_data_2015, 2:cats_data_2016, 3:cats_data_2017}
months = ["January","February","March","April","May","June","July","August","September","October","November","December"]
for i in range(0,4):
    for j in range(0,3):
        if i==0:
            d = cats_data_2014
            y = 2014
        elif i==1:
            d = cats_data_2015
            y = 2015
        elif i==2:
            d = cats_data_2016
            y = 2016
        else:
            d = cats_data_2017
            y = 2017
            
        if j==0:
            cat = "Furniture"
            color = 'teal'
        elif j==1:
            cat = "Office Supplies"
            color = 'slateblue'
        else:
            cat = "Technology"
            color = 'c'
        bar = ax[i][j].bar(months,d.loc[cat], color=color)
        ax[i][j].plot(months,d.loc[cat], '4-.r', label=cat+' '+'Sales'+' - '+str(y))
        if i==3:
            ax[i][j].tick_params('x', labelrotation=45)
            ax[i][j].set_xlabel('Month of the Year', fontsize=14)
        else:
            ax[i][j].set_xticks([])
        if j==0:
            ax[i][j].set_ylabel('Sales in USD ($)', fontsize=14)
            
        ax[i][j].legend(loc='upper left', fontsize=13)
        #autolabel(bar,ax[i][j], rot=90)
```

2. Now let's find the items that are sold together

```python
duplicate_df = data[data['Order ID'].duplicated(keep=False)]
print(duplicate_df.shape)
duplicate_df.head()
```

```python
duplicate_df['Groupped_Products'] = duplicate_df.groupby('Order ID')['Product Name'].transform(lambda x: ','.join(x))
```

```python
duplicate_df = duplicate_df[['Order ID', 'Groupped_Products']].drop_duplicates()
#This line of code will drop the duplicates and will just store the Order ID and Groupped_Products in the duplicate_df

print(duplicate_df.shape)
duplicate_df.head()
```

```python
# Now we need to count the pairs which occur together. We need to use some other libraries for this task

from itertools import combinations
from collections import Counter

count = Counter()

for row in duplicate_df['Groupped_Products']:
    row_list = row.split(',')           #converting the row values to list with comma split
    count.update(Counter(combinations(row_list, 2)))
                 
                 
# print(count)
# count.most_common(10)
#most_common methods only brings the first 10 items.


for k, v in count.most_common(10):
    print(k, v)
```

Even these numbers are not giving us much, because there is not a lot of combined items, based on 16 or 15 items making packages might not be a good idea (Increase in sales might be just as much as you will offer in the discount on packages)
